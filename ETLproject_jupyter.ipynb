```python
import pandas as pd
from sqlalchemy import create_engine
from config import uid, pwd
```

```python
package_csv = "package.csv"
package_df = pd.read_csv(package_csv)
package_df.drop_duplicates("PRODUCTID", inplace=True)
#package_df.set_index("PRODUCTID", inplace=True)
#package_df.head()
```

```python
package_df = package_df.rename(columns={"PRODUCTID": "product_id", "PRODUCTNDC": "product_ndc", "NDCPACKAGECODE": "ndc_package_code", 
                                        "PACKAGEDESCRIPTION": "package_description", "STARTMARKETINGDATE": "start_marketing_date",
                                        "ENDMARKETINGDATE": "end_marketing_date", "NDC_EXCLUDE_FLAG": "ndc_exclude_tag",
                                        "SAMPLE_PACKAGE": "sample_package"})
package_df = package_df[["product_id", "product_ndc", "ndc_package_code", "package_description", "sample_package"]]
package_df.set_index("product_id", inplace=True)
package_df.head()
```

```python
product_csv = "product.csv"
product_df = pd.read_csv(product_csv, encoding = "ISO-8859-1")
product_df.drop_duplicates("PRODUCTID", inplace=True)
#product_df.set_index("PRODUCTID", inplace=True)
#product_df.head()
```

```python
product_df = product_df.rename(columns={"PRODUCTID": "product_id", "PRODUCTNDC": "product_ndc",
                                        "PRODUCTTYPENAME": "product_type_name", "PROPRIETARYNAME": "proprietary_name",                                     
                                        "NONPROPRIETARYNAME": "non_proprietary_name", "DOSAGEFORMNAME": "dosage_form", 
                                        "ROUTENAME": "route", "STARTMARKETINGDATE": "start_marketing_date", 
                                        "ENDMARKETINGDATE": "end_marketing_date", "MARKETINGCATEGORYNAME": "marketing_category",
                                        "APPLICATIONNUMBER": "application_number", "LABELERNAME": "manufacturer",
                                        "SUBSTANCENAME": "substance_name", "ACTIVE_NUMERATOR_STRENGTH": "active_numerator_strength",
                                        "ACTIVE_INGRED_UNIT": "active_ingredient_unit", "PHARM_CLASSES": "pharm_class",
                                        "DEASCHEDULE": "dea_schedule", "NDC_EXCLUDE_FLAG": "ndc_exclude_flag",
                                        "LISTING_RECORD_CERTIFIED_THROUGH": "certified_through"}) 
product_df = product_df[["product_id", "product_ndc", "product_type_name", "proprietary_name", "substance_name", "non_proprietary_name",
                         "dosage_form", "route", "active_numerator_strength", "active_ingredient_unit", "pharm_class", "dea_schedule", "manufacturer",
                         "start_marketing_date", "end_marketing_date"]]
product_df.set_index("product_id", inplace=True)
#product_df.dropna("CVI")
product_df.head()
```

```python
database = "ETLproject"
server = "localhost:5432"
engine = create_engine(f"postgres://{uid}:{pwd}@{server}/{database}")
#conn = engine.connect()
```

```python
engine.table_names()
```

```python
package_df.to_sql(name='package', con=engine, if_exists='append', index=True)
```

```python
product_df.to_sql(name='product', con=engine, if_exists='append', index=True)
```
